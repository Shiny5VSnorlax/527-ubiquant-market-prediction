```python
def missingvalues(train_data):
    count_na = []
    for col in train_data.column_names:
        count_na.append(train_data[col].isna().sum().item())
    a = 0
    for i in range(len(count_na)):
        if count_na[i] == 0:
            print("missing data found in %s"%train_data.column_names[i])
            a = a+1
    if a == 0:
        print("no missing values found")
```
